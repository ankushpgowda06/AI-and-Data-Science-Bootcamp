# Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
np.random.seed(42)
```

### Scatter Plot
-  **When to Use:** To see relationships or correlations between two continuous variables.
-  **Why:** Helps detect patterns, trends, clusters, or outliers.

```python
df_scatter = pd.DataFrame({
    'feature1': np.arange(0, 20),
    'feature2': np.sin(np.arange(0, 20))
})

sns.scatterplot(x='feature1', y='feature2', data=df_scatter)
plt.title("Scatter Plot: Feature1 vs Feature2")
plt.show()
```

### Line Plot
-  **When to Use:** For time series data or ordered sequences.
-  **Why:** Visualizes trends, spikes, or seasonality over time.

```python
df_line = pd.DataFrame({
    'time': np.arange(0, 20),
    'value': np.sin(np.arange(0, 20))
})

sns.lineplot(x='time', y='value', data=df_line)
plt.title("Line Plot: Time vs Value")
plt.show()
```

### Bar Plot
-  **When to Use:** Compare the mean or sum of values across categories.
-  **Why:** Useful for comparing groups clearly and effectively.

```python
df_bar = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D'],
    'values': [23, 45, 56, 12]
})

sns.barplot(x='category', y='values', data=df_bar)
plt.title("Bar Plot: Category vs Values")
plt.show()
```

### Count Plot
-  **When to Use:** To count frequency of categorical variables.
-  **Why:** Quick summary of how often each category appears.

```python
df_count = pd.DataFrame({
    'labels': np.random.choice(['Apple', 'Banana', 'Orange'], size=100)
})

sns.countplot(x='labels', data=df_count)
plt.title("Count Plot: Fruit Frequency")
plt.show()
```

### Box Plot
-  **When to Use:** To see distribution, spread, and outliers of a numeric variable grouped by categories.
-  **Why:** Great for detecting skewness or spotting outliers.

```python
df_box = pd.DataFrame({
    'group': np.random.choice(['X', 'Y', 'Z'], size=100),
    'score': np.random.normal(70, 10, size=100)
})

sns.boxplot(x='group', y='score', data=df_box)
plt.title("Box Plot: Scores by Group")
plt.show()
```

### Violin Plot
-  **When to Use:** Same as box plot, but shows the distribution shape as well.
-  **Why:** Combines boxplot + KDE (distribution estimate) for deeper insight.

```python
df_violin = pd.DataFrame({
    'group': np.random.choice(['Male', 'Female'], size=100),
    'weight': np.random.normal(65, 15, size=100)
})

sns.violinplot(x='group', y='weight', data=df_violin)
plt.title("Violin Plot: Weight by Gender")
plt.show()
```

### Histogram / KDE Plot
-  **When to Use:** To visualize the distribution of a single variable.
-  **Why:**  Helps see skewness, modality (one peak vs two), and spread.

```python
df_hist = pd.DataFrame({
    'height': np.random.normal(170, 10, size=200)
})

sns.histplot(df_hist['height'], kde=True, bins=30)
plt.title("Histogram: Height Distribution")
plt.show()
```

### Heatmap
-  **When to Use:** For correlation matrices, pivot tables, or similarity matrices.
-  **Why:**  Highlights strong/weak correlations visually with color.

```python
df_heat = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})

sns.heatmap(df_heat.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

### Pair Plot
-  **When to Use:** For multivariate exploration of 2+ numeric variables.
-  **Why:**  Helps detect relationships between pairs of features + class separation.

```python
df_pair = pd.DataFrame({
    'feature1': np.random.rand(50),
    'feature2': np.random.rand(50),
    'feature3': np.random.rand(50),
    'label': np.random.choice(['Group1', 'Group2'], size=50)
})

sns.pairplot(df_pair, hue='label')
plt.suptitle("Pair Plot: Feature Interactions", y=1.02)
plt.show()
```

### Cat Plot
-  **When to Use:** To combine categorical + numerical plots (bar, box, strip, etc).
-  **Why:**  Flexible and supports faceting/grouping for complex visualizations.

```python
df_cat = pd.DataFrame({
    'category': np.random.choice(['Low', 'Medium', 'High'], size=100),
    'value': np.random.rand(100) * 100
})

sns.catplot(x='category', y='value', kind='violin', data=df_cat)
plt.title("Cat Plot: Value by Category")
plt.show()
```


# Seaborn Plot Summary Table with Inputs

| Plot Type       | Inputs Required                              | Best For                             | Shows                            |
|------------------|-----------------------------------------------|--------------------------------------|----------------------------------|
| **Scatter Plot** | `x`: numerical, `y`: numerical                | 2 continuous variables                | Correlation/relationship         |
| **Line Plot**    | `x`: time/sequence, `y`: numerical            | Time series or ordered data           | Trends or patterns over time     |
| **Bar Plot**     | `x`: categorical, `y`: numerical              | Group comparisons                     | Means or totals per category     |
| **Count Plot**   | `x`: categorical                              | Frequency of categories               | Distribution of categorical var  |
| **Box Plot**     | `x`: categorical, `y`: numerical              | Distribution & outliers               | Median, IQR, outliers            |
| **Violin Plot**  | `x`: categorical, `y`: numerical              | Distribution shape + box              | KDE + boxplot                    |
| **Histogram**    | `x`: numerical                                | Single variable distribution          | Skew, modality, spread           |
| **Heatmap**      | `data`: 2D matrix or correlation matrix       | Correlation matrix or pivot table     | Strength of relationships        |
| **Pair Plot**    | `data`: dataframe with multiple numerical vars, optional `hue`: categorical | Multivariate EDA                      | Pairwise relationships           |
| **Cat Plot**     | `x`: categorical, `y`: numerical, `kind`: type (e.g., box, bar, etc.) | Flexible categorical analysis         | Box, bar, violin, etc by group   |
